```python
# HW2 Problem2

import pandas as pd
import numpy as np
import requests

# Get stock information

date = "20240318"
url = f'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=json&date={date}&type=ALLBUT0999'
response = requests.get(url)
stock_json = response.json()
stock_data = pd.DataFrame(stock_json['data9'],columns=stock_json['fields9'])
origin_data = stock_data.copy()

# 過濾那些價格有逗號或是當天無開盤資訊導致--的資訊

replace = {',':'','--':np.nan}

stock_data = stock_data.replace(replace,regex=True)
stock_data = stock_data.apply(pd.to_numeric,errors='ignore')

# 過濾當天收盤價不到1000元的雞蛋水餃股

stock_data = stock_data[stock_data.收盤價>=1000]

# 將收盤價轉成numpy array

average = np.array(stock_data['收盤價'])

# 取得當天共有幾家破千元的股票以及平均股價

print("在當天收盤前總共有",len(stock_data),"家公司股價超過千元")
print("當天這些公司的平均股價是", average.mean())

# 取得股王股價

stock_max_money = stock_data.收盤價.max()
stock_name = stock_data[stock_data.收盤價>=stock_max_money]

stock_name.head()
#print(stock_name.證券名稱[838])
#print( stock_name[0])
print("統計到3/18，台股股王是",stock_name.證券名稱[838],"價格為", stock_max_money)
    
```
